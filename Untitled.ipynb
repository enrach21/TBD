```python
i = range(len(train_input.columns))
j = np.random.choice(i, int(len(train_input.columns)*2/3), replace=False)
k = np.random.choice(j, int(len(train_input.columns)/3), replace=False)
K1 = []
K2 = []
K3 = []
OUT1 = []
OUT2 = []
OUT3 = []

for x in i:
    if x not in j:
        K1 += [train_input[train_input.columns[x]]]
        OUT1 += [train_out_scale[train_out_scale.columns[x]]]
for x in j:
    if x not in k:
        K2 += [train_input[train_input.columns[x]]]
        OUT2 += [train_out_scale[train_out_scale.columns[x]]]
for x in k:
    K3 += [train_input[train_input.columns[x]]]
    OUT3 += [train_out_scale[train_out_scale.columns[x]]]
    
K1=np.array(K1).T
K2=np.array(K2).T
K3=np.array(K3).T
OUT1=np.array(OUT1).T
OUT2=np.array(OUT2).T
OUT3=np.array(OUT3).T
```
